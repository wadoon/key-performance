```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 

# this sets up the Matplotlib interactive windows:
# %matplotlib widget

# this changes the default date converter for better interactive plotting of dates:
plt.rcParams['date.converter'] = 'concise'
```

```python
table = pd.read_csv('keyutil.csv', delimiter=";")
table["date"] = table["pwd"].map(lambda x: x.replace("/home/debian/keyrepos/","").replace("/key",""))
plt.bar(table['date'], table['elapsed'], )
plt.xticks(rotation=90)
plt.show()
```

```python
table = pd.read_csv('test.csv', delimiter=";")
table["date"] = table["pwd"].map(lambda x: x.replace("/home/debian/keyrepos/","").replace("/key",""))
table = table.groupby('date')["elapsed"].median(numeric_only=True)
#table.describe()
type(table)
```

```python
plt.grid(True, which="both", axis='y')
plt.bar(table.index,  height=table.values/60.0)
plt.xticks(rotation=90)
plt.xlabel("KeY version on main at given date")
plt.ylabel("Elapsed Time [min]")
plt.show()
```
